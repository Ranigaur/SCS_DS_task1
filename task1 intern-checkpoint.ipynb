```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
msno.matrix(df)
plt.show()


# Check for missing values by column
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])

# Fill missing values or drop if necessary
# Example: Filling Age with the median, and Embarked with the mode
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
# Drop rows with missing 'Fare' or any other important columns (if necessary)
df.dropna(subset=['Fare'], inplace=True)
```

```python
sns.histplot(df['Age'], kde=True, bins=20, color='skyblue')  # KDE (Kernel Density Estimation) helps visualize the distribution curve
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting bar plot for the categorical variable 'Gender'
sns.countplot(x='Sex', data=df, palette='Set2')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a countplot (bar chart) for 'Sex' (categorical data)
sns.countplot(x='sex', data=titanic, palette='Set2')

# Add labels and title
plt.title('Distribution of Gender (Sex)')
plt.xlabel('Gender')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Create a histogram for the 'Age' column with KDE (Kernel Density Estimate)
sns.histplot(titanic['age'].dropna(), kde=True, bins=20, color='skyblue')

# Add labels and title
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
