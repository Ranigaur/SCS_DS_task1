```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
df=pd.read_csv("tested.csv")
print(df)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python


msno.matrix(df)
plt.show()


# Check for missing values by column
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])

# Fill missing values or drop if necessary
# Example: Filling Age with the median, and Embarked with the mode
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
# Drop rows with missing 'Fare' or any other important columns (if necessary)
df.dropna(subset=['Fare'], inplace=True)
```

histogram  for cotinuous variable 'AGE':

```python

sns.histplot(df['Age'], kde=True, bins=20, color='skyblue')  # KDE (Kernel Density Estimation) helps visualize the distribution curve
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

 Barplot for continuous variable 'GENDER' :

```python
# Plotting bar plot for the categorical variable 'Gender'
sns.countplot(x='Sex', data=df, palette='Set2')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

combining both above graphs for EDA:

```python
# For continuous variables
sns.histplot(df['Age'], kde=True, bins=20, color='lightgreen')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# For categorical variables
sns.countplot(x='Sex', data=df, palette='Pastel1')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Univariate analysis(single variable analysis):

```python
# Survival distribution
sns.countplot(x='Survived', data=df)
plt.title('Survival Distribution')
plt.show()

# Categorical features distribution
sns.countplot(x='Pclass', data=df)
plt.title('Class Distribution')
plt.show()

sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()

sns.countplot(x='Embarked', data=df)
plt.title('Embarked Distribution')
plt.show()
```

bivariate analysis

```python
# Survival by Sex
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Sex')
plt.show()

# Survival by Pclass
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Pclass')
plt.show()

# Survival by Embarked
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival by Embarked')
plt.show()

# Survival by Family size (SibSp + Parch)
df['FamilySize'] = df['SibSp'] + df['Parch']
sns.countplot(x='FamilySize', hue='Survived', data=df)
plt.title('Survival by Family Size')
plt.show()
```

Correlation Matrix for Numerical Features


```python
# Only select numeric columns for analysis
numeric_df = df.select_dtypes(include=[np.number])

# Then you can perform operations like correlation analysis
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

 Outlier Detection (Boxplot for Numerical Features)



```python
# Boxplot for detecting outliers in Fare and Age
sns.boxplot(x=df['Fare'])
plt.title('Fare Outliers')
plt.show()

sns.boxplot(x=df['Age'])
plt.title('Age Outliers')
plt.show()
```

Multivariate Analysis (Interactions Between Multiple Variables)


```python
# Pairplot with hue='Survived' (to show survival based on different features)
sns.pairplot(df, hue='Survived', vars=['Age', 'Fare', 'SibSp', 'Parch'])
plt.show()
```

Feature Engineering



```python
# Create new features like FamilySize
df['FamilySize'] = df['SibSp'] + df['Parch']
df['IsAlone'] = 1 # Assuming everyone is alone by default
df['IsAlone'].loc[df['FamilySize'] > 0] = 0 # Update if they have family

# Create age groups
bins = [0, 12, 18, 30, 50, 100]
labels = ['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Check the new features
print(df[['FamilySize', 'IsAlone', 'AgeGroup']].head())
```

Visualizing Feature Engineering



```python
# Survival by Family Size
sns.countplot(x='FamilySize', hue='Survived', data=df)
plt.title('Survival by Family Size')
plt.show()

# Survival by Age Group
sns.countplot(x='AgeGroup', hue='Survived', data=df)
plt.title('Survival by Age Group')
plt.show()

# Survival by IsAlone
sns.countplot(x='IsAlone', hue='Survived', data=df)
plt.title('Survival by IsAlone')
plt.show()
```

Summary of Insights
After performing these analyses

Survival by class: The survival rate in 1st class is higher than in 3rd class.

Survival by gender: Women have a higher survival rate than men.

Age: Younger people, especially children, tend to have a higher survival rate.

Family Size: Passengers traveling with family members had a higher survival rate than those traveling alone.
